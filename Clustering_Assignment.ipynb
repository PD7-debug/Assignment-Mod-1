# Assignment: Clustering

### Theory Question 1: What is unsupervised learning in the context of machine learning

**Answer:**  
Unsupervised learning is a type of machine learning where the model is not given any labeled data. Instead, it tries to learn the underlying structure or distribution in the data to discover hidden patterns or groupings. Clustering is a common unsupervised learning task.

### Theory Question 2: How does K-Means clustering algorithm work

**Answer:**  
K-Means clustering works by initializing `k` cluster centers randomly. It then performs two steps iteratively:
1. Assign each data point to the nearest cluster center.
2. Update each cluster center to the mean of its assigned points.
This process continues until the assignments do not change significantly.

### Theory Question 3: Explain the concept of a dendrogram in hierarchical clustering

**Answer:**  
A dendrogram is a tree-like diagram that records the sequences of merges or splits in hierarchical clustering. Each level of the tree represents a merging of clusters based on a distance or similarity metric, allowing one to visually assess how clusters are formed.

### Theory Question 4: What is the main difference between K-Means and Hierarchical Clustering

**Answer:**  
K-Means partitions the data into `k` flat clusters using distance to centroids, while hierarchical clustering builds a tree of clusters through successive merges or splits, without needing a predefined number of clusters.

### Theory Question 5: What are the advantages of DBSCAN over K-Means

**Answer:**  
DBSCAN has advantages over K-Means such as:
- It does not require specifying the number of clusters beforehand.
- It can find clusters of arbitrary shape.
- It identifies and handles noise (outliers) automatically.

### Theory Question 6: When would you use Silhouette Score in clustering

**Answer:**  
Silhouette Score measures how similar a data point is to its own cluster compared to other clusters. It helps evaluate the quality of clustering and is used to determine the optimal number of clusters.

### Theory Question 7: What are the limitations of Hierarchical Clustering

**Answer:**  
Hierarchical clustering has limitations like:
- High computational complexity for large datasets.
- Irreversible merging/splitting steps.
- Sensitive to noise and outliers.

### Theory Question 8: Why is feature scaling important in clustering algorithms like K-Means

**Answer:**  
Feature scaling ensures that all features contribute equally to the distance calculations used by clustering algorithms like K-Means, which rely heavily on Euclidean distance.

### Theory Question 9: How does DBSCAN identify noise points

**Answer:**  
DBSCAN identifies noise points as those that do not meet the density requirements (i.e., fewer than `min_samples` points within `eps` distance). These are labeled as -1.

### Theory Question 10: Define inertia in the context of K-Means

**Answer:**  
Inertia is the sum of squared distances between each data point and the centroid of its assigned cluster. It measures how tightly the clusters are packed.

### Theory Question 11: What is the elbow method in K-Means clustering

**Answer:**  
The elbow method plots the inertia against the number of clusters. The 'elbow' point where the rate of decrease sharply changes suggests the optimal number of clusters.

### Theory Question 12: Describe the concept of "density" in DBSCAN

**Answer:**  
In DBSCAN, density refers to the number of points within a given radius (`eps`) from a point. A dense region has more than `min_samples` within that radius.

### Theory Question 13: Can hierarchical clustering be used on categorical data

**Answer:**  
Yes, with proper distance metrics (like Gower distance), hierarchical clustering can be applied to categorical data, although it's not as straightforward as with numeric data.

### Theory Question 14: What does a negative Silhouette Score indicate

**Answer:**  
A negative Silhouette Score means that a sample is assigned to the wrong cluster and is closer to a neighboring cluster than its own.

### Theory Question 15: Explain the term "linkage criteria" in hierarchical clustering

**Answer:**  
Linkage criteria determine how the distance between clusters is calculated. Examples include:
- Single linkage: minimum distance between points.
- Complete linkage: maximum distance.
- Average linkage: average pairwise distance.

### Theory Question 16: Why might K-Means clustering perform poorly on data with varying cluster sizes or densities

**Answer:**  
K-Means can perform poorly when:
- Clusters are of different sizes or densities.
- Clusters are not spherical or are overlapping.
- Initial centroids are poorly chosen.

### Theory Question 17: What are the core parameters in DBSCAN, and how do they influence clustering

**Answer:**  
Core parameters of DBSCAN:
- `eps`: The radius within which to search for neighboring points.
- `min_samples`: Minimum number of points to form a dense region.
These influence how DBSCAN identifies clusters and noise.

### Theory Question 18: How does K-Means++ improve upon standard K-Means initialization

**Answer:**  
K-Means++ initializes centroids to be distant from each other, which improves convergence and reduces chances of poor clustering results compared to random initialization.

### Theory Question 19: What is agglomerative clustering

**Answer:**  
Agglomerative clustering is a bottom-up approach where each data point starts as its own cluster, and pairs of clusters are merged based on distance until one cluster remains or a stopping condition is met.

### Theory Question 20: What makes Silhouette Score a better metric than just inertia for model evaluation?

**Answer:**  
Silhouette Score considers both cohesion and separation of clusters, whereas inertia only considers compactness. Hence, Silhouette Score provides a more comprehensive evaluation.

### Practical Question 31: Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Step 1: Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# Step 2: Fit K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Plot decision boundaries
h = 0.05  # mesh step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each point in the mesh grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering with Decision Boundaries')
plt.legend()
plt.show()
```

### Practical Question 34: Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
data = load_breast_cancer()
X = data.data

# Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# Calculate inertia for K = 2 to 6
inertias = []
k_values = range(2, 7)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot inertia values
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o')
plt.title('K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### Practical Question 44: Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

# Load the Iris dataset
data = load_iris()
X = data.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette scores for each sample
silhouette_vals = silhouette_samples(X, labels)

# Plot the silhouette scores
plt.figure(figsize=(8, 6))
y_lower = 10
for i in np.unique(labels):
    ith_cluster_silhouette_values = silhouette_vals[labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values, label=f'Cluster {i}')
    y_lower = y_upper + 10

plt.axvline(x=np.mean(silhouette_vals), color="red", linestyle="--", label="Average Silhouette Score")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.title("Silhouette Plot for KMeans Clustering on Iris Dataset")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Load Wine dataset
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
X['cluster'] = kmeans.fit_predict(X).astype(str)  # Convert cluster labels to string

# Visualize clusters using seaborn pairplot (first 4 features)
sns.pairplot(X, hue='cluster', vars=wine.feature_names[:4], palette='viridis')
plt.suptitle('KMeans Clustering on Wine Dataset', y=1.02)
plt.show()
```

### Practical Question 47: Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.8, random_state=42)
X = np.vstack([X, np.random.uniform(low=-10, high=10, size=(50, 2))])

dbscan = DBSCAN(eps=1.0, min_samples=5)
labels = dbscan.fit_predict(X)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = (labels == -1).sum()

print(f'Clusters: {n_clusters}, Noise points: {n_noise}')
```

### Practical Question 48: Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()
X = digits.data

# Dimensionality reduction using t-SNE
X_embedded = TSNE(n_components=2, random_state=42).fit_transform(X)

# Agglomerative Clustering
clusters = AgglomerativeClustering(n_clusters=10).fit_predict(X_embedded)

# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=clusters, cmap='tab10', s=50, alpha=0.8)
plt.title('Agglomerative Clustering of Digits Dataset (t-SNE)')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.grid(True)
plt.show()
```
